# 12. Find All Substrings of a String
## Problem Statement:
Write a function to return all substrings of a given string.

## Example 1:
**Input:** `"abc"`  
**Output:** `["a", "b", "c", "ab", "bc", "abc"]`

## Example 2:
**Input:** `"xyz"`  
**Output:** `["x", "y", "z", "xy", "yz", "xyz"]`


```python

def substrings(lst):
    substrings = []
    n = len(lst)

    for x in range(1, n + 1):
        for i in range(n - x + 1):
            substrings.append(lst[i:i+x])

    return substrings
lst="xyz"
print(substrings(lst))
```

# 14. Find the Longest Common Prefix of a List of Strings
## Problem Statement:
Write a function to find the longest common prefix among a list of strings.

## Example 1:
**Input:** `["flower", "flow", "flight"]`  
**Output:** `"fl"`

## Example 2:
**Input:** `["dog", "racecar", "car"]`  
**Output:** `""`


```python
def longest_common_prefix(strs):
    if not strs:
        return ""

    prefix = strs[0]

    for x in strs[1:]:
        while not x.startswith(prefix):
            prefix = prefix[:-1]
            if prefix == "":
                return ""

    return prefix
strs = ["flower", "flow", "flight"]
print(longest_common_prefix(strs))
```

# 16. Find the Second Most Frequent Character in a String
## Problem Statement:
Write a function to find the second most frequent character in a string.

## Example 1:
**Input:** `"banana"`  
**Output:** `"n"`

## Example 2:
**Input:** `"mississippi"`  
**Output:** `"s"`


```python
def most_frequently(str1):
    new_dict = {}
    for x in str1:
        new_dict[x] = new_dict.get(x, 0) + 1

    new_dict1 = sorted(new_dict.items(), key=lambda y: y[1], reverse=True)

    if len(new_dict1) >= 2:
        return new_dict1[1][0]
    else:
        return None

str1 = "mississippi"
print(most_frequently(str1))
```

```python
def dublicate_remv(str):
    vowels = set("abcdefghijklmnopqrstuvwxyz")
    new_lst = ""
    for x in str:
        if x in vowels:
            if x not in new_lst:
                new_lst += x
    return new_lst

str = "mississippi"
print(dublicate_remv(str))
```

# 22. Implement Run-Length Encoding
## Problem Statement:
Write a function to perform run-length encoding on a string.

## Example 1:
**Input:** `"aaabbc"`  
**Output:** `"3a2b1c"`

## Example 2:
**Input:** `"wwwwaaadexxxxxx"`  
**Output:** `"4w3a1d1e6x"`


```python
def run_length_encoding(str1):
    result = ""
    count = 1

    for i in range(1, len(str1)):
        if str1[i] == str1[i-1]:
            count += 1
        else:
            result += str1[i-1] + str(count)
            count = 1

    result += str1[-1] + str(count)

    return result
str1="wwwwaaadexxxxxx"
print(run_length_encoding(str1))
```

# 23. Find All Permutations of a String
## Problem Statement:
Write a function to generate all permutations of a given string.

## Example 1:
**Input:** `"abc"`  
**Output:** `["abc", "acb", "bac", "bca", "cab", "cba"]`

## Example 2:
**Input:** `"ab"`  
**Output:** `["ab", "ba"]`


```python
def permutation(str1):

    if len(str1) == 1:
        return [str1]

    result = []
    for i in range(len(str1)):
        current_ch = str1[i]
        remaining = str1[:i] + str1[i+1:]
        for x in permutation(remaining):
            result.append(current_ch + x)
    return result


str1="abc"
print(permutation(str1))

```
